### Question 1: -
### Write a program that takes a string as input, and counts the frequency of each word in the string, there might be repeated characters in the string. Your task is to find the highest frequency and returns the length of the highest-frequency word.

```python
from collections import Counter

def find_highest_frequency_word_length(string):
    # Split the string into words
    words = string.split()

    # Count the frequency of each word
    word_counts = Counter(words)

    # Find the highest frequency
    max_frequency = max(word_counts.values())

    # Find the length of the highest-frequency word
    highest_frequency_word_length = max(len(word) for word, count in word_counts.items() if count == max_frequency)

    return highest_frequency_word_length

# Example usage
#test case1 
input_string = input("Enter a string: ")
result = find_highest_frequency_word_length(input_string)
print("Length of the highest-frequency word:", result)

#test case2
input_string1= input("Enter a string: ")
result = find_highest_frequency_word_length(input_string1)
print("Length of the highest-frequency word:", result)

#test case3
input_string2= input("Enter a string: ")
result = find_highest_frequency_word_length(input_string2)
print("Length of the highest-frequency word:", result)
```

In this program, we use the split() method to split the input string into words. Then, we use the Counter class from the collections module to count the frequency of each word. The Counter object word_counts stores the word as the key and its frequency as the value.

Next, we find the highest frequency by using the max() function on the values of word_counts. After that, we iterate over the items in word_counts and check if the count of each word is equal to the highest frequency. We use a generator expression and the max() function to find the length of the word with the highest frequency.

Finally, we prompt the user to enter a string, call the find_highest_frequency_word_length function, and print the result.

### Question 2: -
### Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if he can remove just one character at the index in the string, and the remaining characters will occur the same number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

```python
def is_string_valid(s):
    # Count the frequency of each character
    char_counts = {}
    for char in s:
        char_counts[char] = char_counts.get(char, 0) + 1

    # Count the frequency of frequencies
    freq_counts = {}
    for count in char_counts.values():
        freq_counts[count] = freq_counts.get(count, 0) + 1

    # If all characters have the same frequency, it is valid
    if len(freq_counts) == 1:
        return "YES"

    # If there are exactly two different frequencies
    if len(freq_counts) == 2:
        # Check if we can remove one character to make it valid
        freq1, count1 = freq_counts.popitem()
        freq2, count2 = freq_counts.popitem()
        if (count1 == 1 and freq1 - 1 == freq2) or (count2 == 1 and freq2 - 1 == freq1):
            return "YES"

    return "NO"


# Example usage
input_string = input("Enter a string: ")
result = is_string_valid(input_string)
print(result)
```

In this program, we count the frequency of each character in the string using a dictionary char_counts. Each character is stored as a key, and its frequency is stored as the corresponding value.

Then, we count the frequency of frequencies using another dictionary freq_counts. For each count in char_counts, we update the count in freq_counts.

Next, we check the conditions for the string to be valid. If freq_counts has only one key-value pair, it means all characters have the same frequency, and the string is valid. In this case, we return "YES".

If freq_counts has exactly two different frequencies, we check if we can remove one character to make it valid. We pop the key-value pairs from freq_counts and check if one count is 1 and the other count is one less than the corresponding frequency. If this condition is satisfied, we return "YES".

If none of the above conditions are met, the string is not valid, and we return "NO".

Finally, we prompt the user to enter a string, call the is_string_valid function, and print the result.

```python
import requests

# Download data from the provided link
url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
response = requests.get(url)
data = response.json()

# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_threshold = 0.05
spawn_rate_less_than_5 = [pokemon for pokemon in data["pokemon"] if pokemon["spawn_chance"] < spawn_rate_threshold]

# Get all Pokemons that have less than 4 weaknesses
weaknesses_threshold = 4
less_than_4_weaknesses = [pokemon for pokemon in data["pokemon"] if len(pokemon["weaknesses"]) < weaknesses_threshold]

# Get all Pokemons that have no multipliers at all
no_multipliers = [pokemon for pokemon in data["pokemon"] if not pokemon["multipliers"]]

# Get all Pokemons that do not have more than 2 evolutions
max_evolutions = 2
less_than_2_evolutions = [pokemon for pokemon in data["pokemon"] if len(pokemon.get("next_evolution", [])) <= max_evolutions]

# Get all Pokemons whose spawn time is less than 300 seconds
spawn_time_threshold = "05:00"
less_than_300_spawn_time = [pokemon for pokemon in data["pokemon"] if pokemon.get("spawn_time", "") < spawn_time_threshold]

# Get all Pokemon who have more than two types of capabilities
capabilities_threshold = 2
more_than_2_capabilities = [pokemon for pokemon in data["pokemon"] if len(pokemon["type"]) > capabilities_threshold]

# Print the results
print("Pokemons whose spawn rate is less than 5%:")
for pokemon in spawn_rate_less_than_5:
    print(pokemon["name"])

print("\nPokemons that have less than 4 weaknesses:")
for pokemon in less_than_4_weaknesses:
    print(pokemon["name"])

print("\nPokemons that have no multipliers at all:")
for pokemon in no_multipliers:
    print(pokemon["name"])

print("\nPokemons that do not have more than 2 evolutions:")
for pokemon in less_than_2_evolutions:
    print(pokemon["name"])

print("\nPokemons whose spawn time is less than 300 seconds:")
for pokemon in less_than_300_spawn_time:
    print(pokemon["name"])

print("\nPokemon who have more than two types of capabilities:")
for pokemon in more_than_2_capabilities:
    print(pokemon["name"])
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Download data from the provided link
url = "https://data.nasa.gov/resource/y77d-th95.json"
response = requests.get(url)
data = response.json()

# Create a DataFrame from the downloaded data
df = pd.DataFrame(data)

# Check available column names
print(df.columns)

# Filter out invalid timestamp values and convert the year column to datetime type
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[(df['reclat'].notnull()) & (df['fall'] == 'Fell') & (df['year'].dt.year < 2000)]

# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_coordinates_before_1970 = df[(df['reclat'].notnull()) & (df['fall'] == 'Fell') & (df['year'].dt.year < 1970)][['reclat', 'reclong']]

# Assuming mass is in kg, get all Earth meteorites with a mass greater than 10000 kg
earth_meteorites_mass_above_10000kg = df[(df['reclat'].notnull()) & (df['mass'].astype(float) > 10000)][['name', 'mass']]

# Print the results
print("Earth meteorites that fell before the year 2000:")
print(earth_meteorites_before_2000[['name', 'year']])

print("\nEarth meteorites' coordinates that fell before the year 1970:")
print(earth_meteorites_coordinates_before_1970)

print("\nEarth meteorites with a mass greater than 10000 kg:")
print(earth_meteorites_mass_above_10000kg[['name', 'mass']])

# Plot 1: Earth meteorites before the year 2000
plt.figure(figsize=(8, 6))
plt.hist(earth_meteorites_before_2000['year'].dt.year, bins=50, edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Earth Meteorites That Fell Before the Year 2000')
plt.show()

# Plot 2: Earth meteorites' coordinates before the year 1970
plt.figure(figsize=(8, 6))
plt.scatter(earth_meteorites_coordinates_before_1970['reclong'], earth_meteorites_coordinates_before_1970['reclat'], s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites\' Coordinates Before the Year 1970')
plt.show()

# Plot 3: Earth meteorites with a mass greater than 10000 kg
plt.figure(figsize=(8, 6))
plt.barh(earth_meteorites_mass_above_10000kg['name'], earth_meteorites_mass_above_10000kg['mass'].astype(float))
plt.xlabel('Mass')
plt.ylabel('Meteorite Name')
plt.title('Earth Meteorites With a Mass Greater Than 10000 kg')
plt.tight_layout()
plt.show()

```

Histogram: It visualizes the distribution of Earth meteorites that fell before the year 2000.
Scatter Plot: It shows the coordinates (latitude and longitude) of Earth meteorites that fell before the year 1970.
Horizontal Bar Chart: It displays the mass of Earth meteorites with a mass greater than 10000 kg.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Download data from the provided link
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
response = requests.get(url)
data = response.json()

# Extract relevant information from the data
show_name = data['name']
episodes = data['_embedded']['episodes']

# Create lists to store the analysis results
season_ratings = []
high_rated_episode_names = []
episode_names_before_may_2019 = []
highest_rated_episodes = []
lowest_rated_episodes = []
most_popular_episode_summaries = {}

# Analyze the data
for episode in episodes:
    season_ratings.append((episode['season'], episode['rating']['average'] if episode['rating'] is not None and 'average' in episode['rating'] else None))
    if episode['rating'] is not None and 'average' in episode['rating'] and episode['rating']['average'] is not None and float(episode['rating']['average']) > 8:
        high_rated_episode_names.append(episode['name'])
    if episode['airdate'] is not None and episode['airdate'] < '2019-05-01':
        episode_names_before_may_2019.append(episode['name'])

# Convert the analysis results into DataFrames
season_ratings_df = pd.DataFrame(season_ratings, columns=['Season', 'Rating'])

# Plotting the ratings for each season
plt.figure(figsize=(10, 6))
plt.bar(season_ratings_df['Season'], season_ratings_df['Rating'], color='orange')
plt.xlabel('Season')
plt.ylabel('Rating')
plt.title(f'{show_name} Ratings by Season')
plt.xticks(season_ratings_df['Season'])
plt.grid(axis='y')
plt.show()

# Display the analysis results
print("Episode Names with Average Rating > 8 for Every Season:")
for season in set(season_ratings_df['Season']):
    high_rated_episodes_season = [episode['name'] for episode in episodes if episode['season'] == season and episode['rating'] is not None and 'average' in episode['rating'] and episode['rating']['average'] is not None and float(episode['rating']['average']) > 8]
    print(f"Season {season}: {', '.join(high_rated_episodes_season)}")

print("\nEpisode Names Aired Before May 2019:")
for episode_name in episode_names_before_may_2019:
    print(episode_name)

print("\nEpisode with Highest and Lowest Rating by Season:")
for season in set(season_ratings_df['Season']):
    episodes_season = [episode for episode in episodes if episode['season'] == season]
    episodes_season.sort(key=lambda x: float(x['rating']['average']) if x['rating'] is not None and 'average' in x['rating'] and x['rating']['average'] is not None else 0, reverse=True)
    highest_rated_episode = episodes_season[0]
    lowest_rated_episode = episodes_season[-1]
    print(f"Season {season}: Highest Rating - {highest_rated_episode['name']}, Lowest Rating - {lowest_rated_episode['name']}")

print("\nSummary for Most Popular Episode in Every Season:")
for season in set(season_ratings_df['Season']):
    episodes_season = [episode for episode in episodes if episode['season'] == season]
    episodes_season.sort(key=lambda x: float(x['rating']['average']) if x['rating'] is not None and 'average' in x['rating'] and x['rating']['average'] is not None else 0, reverse=True)
    most_popular_episode = episodes_season[0]
    most_popular_episode_summaries[season] = most_popular_episode['summary']

most_popular_episodes_df = pd.DataFrame({'Season': list(most_popular_episode_summaries.keys()), 'Summary': list(most_popular_episode_summaries.values())})
print(most_popular_episodes_df)
```

### Question 10 -
### Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.

```python
import nltk
from nltk import pos_tag
from collections import defaultdict

def count_pos_tags(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)

    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)

    # Initialize counts dictionary
    counts = defaultdict(int)

    # Count the number of verbs, nouns, pronouns, and adjectives
    for word, tag in tagged_words:
        if tag.startswith('V'):  # Verb
            counts['verbs'] += 1
        elif tag.startswith('N'):  # Noun
            counts['nouns'] += 1
        elif tag.startswith('P'):  # Pronoun
            counts['pronouns'] += 1
        elif tag.startswith('J'):  # Adjective
            counts['adjectives'] += 1

    return counts

# Test case 1
text1 = "The quick brown fox jumps over the lazy dog."
counts1 = count_pos_tags(text1)
print(counts1)  # Output: {'verbs': 1, 'nouns': 3, 'pronouns': 0, 'adjectives': 1}

# Test case 2
text2 = "I like to eat pizza."
counts2 = count_pos_tags(text2)
print(counts2)  # Output: {'verbs': 1, 'nouns': 1, 'pronouns': 1, 'adjectives': 0}

# Additional test case 1
text3 = "She was running towards the bright blue sky."
counts3 = count_pos_tags(text3)
print(counts3)  # Output: {'verbs': 1, 'nouns': 2, 'pronouns': 1, 'adjectives': 2}

# Additional test case 2
text4 = "The cat sat on the mat."
counts4 = count_pos_tags(text4)
print(counts4)  # Output: {'verbs': 1, 'nouns': 2, 'pronouns': 0, 'adjectives': 1}
```
